```python
# --- Load Agents --- #
from agents.agent_deepqn.agent import DeepQAgent
from agents.agent_reinforce.agent import REINFORCEAgent

# --- Load Environments --- #
from environment.grid import GridEnv

# --- Load Necessary --- #
from collections import deque
import numpy as np
from utilities.helper import flatten
from utilities.profiler import profile

# --- Load Training --- #
from experiments.experiment import Experiment

%load_ext autoreload
%autoreload 2
```

```python
env = GridEnv(num_agent = 1, grid_size = 5, fixed_start = False, fixed_goals = True, prob_right_direction = 1, render_board = False)
```

```python
max_t = 100
n_episodes =1500
```

```python
agents_dqn = [DeepQAgent(env.state_space, env.action_space.n, seed=1) for _ in range(env.num_agent)]
```

```python
agents_rei = [REINFORCEAgent(env.state_space, env.action_space.n, seed=1) for _ in range(env.num_agent)]
```

```python
experiments_dqn = [Experiment(name="DQN Exp {}".format(1), environment = env, agents=agents_dqn, max_t=max_t, num_episodes=n_episodes, goal = 0., save_states_every=10, experiment_num=1)]
```

```python
experiments_rei = [Experiment(name="REI Exp {}".format(1), environment = env, agents=agents_rei, max_t=max_t, num_episodes=n_episodes, goal = 0., save_states_every=10, experiment_num=1)]
```

```python
for experiment in experiments_dqn:
    score_history, state_history = experiment.run()
    experiment.save(score_history, state_history)
```

```python
for experiment in experiments_rei:
    score_history, state_history = experiment.run()
    experiment.save(score_history, state_history)
```
